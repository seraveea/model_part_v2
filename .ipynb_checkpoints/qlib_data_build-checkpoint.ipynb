```python
import tushare as ts
import numpy as np
from tqdm import tqdm
import time
import re
# ts.set_token('2d7d464e46a43f8e58efcac32727985d0c31de986a87c2e5053d8761')
# pro = ts.pro_api()
# ts.pro_bar(ts_code='000001.SZ', adj='hfq', start_date='20180101', end_date='20181011',adjfactor=True)
import akshare as ak
data = np.load('data/csi300_stock_index.npy',allow_pickle=True)
symbol_list = list(data.tolist().keys())
# SH999999,SH999998有问题
symbol_list = symbol_list[412:]
symbol_list
```

```python
for symbol in tqdm(symbol_list):
    num_symbol = re.sub("[^0-9]", "", symbol)
    df = ak.stock_zh_a_hist(symbol=num_symbol, period="daily", start_date="20070101", end_date='20221231', adjust="hfq")
    df = df.rename(columns={'日期':'date','开盘':'open','收盘':'close','最高':'high','最低':'low','成交量':'volume','成交额':'amount'})
    df.to_csv('../csv_data/'+symbol+'.csv', index=False)
    time.sleep(3)
    
```
