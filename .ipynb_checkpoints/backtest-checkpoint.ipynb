```python
# use bs4 to get all symbol
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup as bf
import time
import re
import json
from tqdm import tqdm
import time
import random
import requests
import sys
```

```python
url = "https://companiesmarketcap.com/usa/largest-companies-in-the-usa-by-market-cap/"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/67.0.3396.99 Safari/537.36'}

response = Request(url, headers=headers)
html = urlopen(response)
obj = bf(html.read(), 'html.parser')
company_code = obj.find_all(attrs={'class':'company-code'})
code_list = []
for code in company_code:
    code_list.append(code.text)
```

```python
company_name = obj.find_all(attrs={'class':'company-name'})
name_list = []
for name in company_name:
    name_list.append(name.text)
```

```python
name_list
us_top100_dict = {name_list[i]: code_list[i] for i in range(len(name_list))}
```

```python
us_top100_dict
# update in 4/22 2023
with open('data/us_top100.json', 'w') as fp:
    json.dump(us_top100_dict, fp)
```

```python
import pandas as pd
temp=pd.read_pickle("data/alpha360.pkl")
temp
```

```python
# VWAP is not existed
first_index = df.index.tolist()
index_list = []
for i in first_index[60:]:
    for j in us_top100_dict.values():
        index_list.append([i,j])
index_df = pd.DataFrame(index_list, columns=["datetime", "instrument"])
index = pd.MultiIndex.from_frame(index_df)

all_col=[]
for i in ['VOLUME','CLOSE','HIGH','OPEN','VWAP']:
    all_col.extend([i+'_'+str(x) for x in range(60)])
multi_col = [['feature',x] for x in all_col]
multi_col.append(['label', 'score'])
col_df = pd.DataFrame(multi_col)
col = pd.MultiIndex.from_frame(col_df)

alpha360_us = pd.DataFrame(index=index, columns=col)
alpha360_us
```

```python
for i in tqdm(first_index[60:]):
    cur_num = first_index.index(i)
    for sym in us_top100_dict.values():
        for offset in range(60):
            alpha360_us["feature", "VOLUME_"+str(offset)].loc[i,sym]=df.loc[first_index[cur_num-offset]]["Volume",sym]
    
    
```
